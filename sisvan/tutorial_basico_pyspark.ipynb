```python
import pyspark.sql.functions as f
import pyspark.sql.types as t
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("meu_projeto")
    .master("local[*]") 
    .config("spark.sql.shuffle.partitions", "8")  
    .config("spark.driver.memory", "4g")          
    .config("spark.executor.memory", "4g")        
    .config("spark.sql.execution.arrow.pyspark.enabled", "true")  
    .getOrCreate()
)
```

## Criando uma base de dados fictícia
Vamos criar uma lista de dicionários representando funcionários de uma empresa e convertê-la em um DataFrame do PySpark.

```python
dados = [
    {"id": 1, "nome": "Ana", "idade": 29, "departamento": "Vendas", "salario": 3500},
    {"id": 2, "nome": "Bruno", "idade": 34, "departamento": "TI", "salario": 5000},
    {"id": 3, "nome": "Carla", "idade": 25, "departamento": "Marketing", "salario": 3200},
    {"id": 4, "nome": "Daniel", "idade": 40, "departamento": "TI", "salario": 6000},
    {"id": 5, "nome": "Eduarda", "idade": 28, "departamento": "Vendas", "salario": 3700}
]

df = spark.createDataFrame(dados)
df.show()
```

## Explorando os dados
Vamos verificar o schema (estrutura) e algumas estatísticas básicas.

```python
df.printSchema()
df.describe().show()
```

## Selecionando colunas e filtrando dados

```python
df.select('nome', 'departamento').show()

# Filtrar funcionários com salário acima de 4000
df.filter(df['salario'] > 4000).show()
```

## Agregações e agrupamentos

```python
from pyspark.sql import functions as F

# Média salarial por departamento
df.groupBy('departamento').agg(F.avg('salario').alias('media_salario')).show()
```

## Adicionando e modificando colunas

```python
# Criar uma nova coluna com aumento de 10%
df_com_aumento = df.withColumn('salario_com_aumento', df['salario'] * 1.1)
df_com_aumento.show()
```
